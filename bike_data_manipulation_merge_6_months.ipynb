## Extract required information from dataset
### To predict number of bikes in each station in near future
- Number of arrival bike for each station
- Number of depature bike for each station
- Number of total bike for each station


```python
# Extract the daily data by the date in the dataset.

newDataset_six_month = (pd.to_datetime(dataset_six_month['End Date'], dayfirst = True)
        .dt.floor('d')
        .reset_index(name = 'timestamp'))
        

newDataset_six_month['EndStation Id'] = dataset_six_month['EndStation Id'].values
newDataset_six_month['StartStation Id'] = dataset_six_month['StartStation Id'].values

newDataset_six_month
```

```python
# Extract number of bikes arriving to the station
In = newDataset_six_month.groupby(['timestamp','EndStation Id'])['EndStation Id'].count().reset_index(name ='In')
In = In[(In['timestamp'] >= '2016-04-01') & ((In['timestamp'] <= '2016-10-31')) & (In['EndStation Id'] >= 1)]
In
```

```python
# Extract number of bikes leaving to the station
Out = newDataset_six_month.groupby(['timestamp','StartStation Id'])['StartStation Id'].count().reset_index(name ='Out')
Out = Out[(Out['timestamp'] >= '2016-04-01') & ((Out['timestamp'] <= '2016-10-31')) & (Out['StartStation Id'] >= 1)]
Out
```

```python
# combine dataframe : In and Out 

InOut_six_month = Out[['timestamp']].copy()
InOut_six_month['In'] = In[['In']].copy()        
InOut_six_month['Out'] = Out[['Out']].copy()
InOut_six_month['Station Id'] =  In[['EndStation Id']].copy()
InOut_six_month['Total'] = InOut_six_month['In'] - InOut_six_month['Out']

InOut_six_month = InOut_six_month.fillna(0)
InOut_six_month['In'] = InOut_six_month['In'].astype(int)
InOut_six_month['Out'] = InOut_six_month['Out'].astype(int)
InOut_six_month['Station Id'] = InOut_six_month['Station Id'].astype(int)
InOut_six_month = InOut_six_month[(InOut_six_month['Station Id'] >= 1) & (InOut_six_month['timestamp'] >= '2016-04-01')]

InOut_six_month
```

```python
df_by_day = condition_df.resample('D').sum()
df_by_day
```

```python
# To calculate the average values for conditions

num_hours = 24

average_df = df_by_day[['is_holiday', 'is_weekend']].copy()
average_df['avg_temp'] = df_by_day['t1']/num_hours
average_df['avg_temp_feel'] = df_by_day['t2']/num_hours
average_df['avg_humidity'] = df_by_day['hum']/num_hours
average_df['avg_windSpeed'] = df_by_day['wind_speed']/num_hours
average_df['date'] = condition_df['date']
```

### Add Longitude and Latitude for the station location

```python
location = pd.read_csv("Cycle hire docking stations.csv")
```

```python
merged_df_six_month['longitude'] = 0
merged_df_six_month['latitude'] = 0
```

```python
cnt = 0
loc_cnt = 0

for cnt in range(0, len(merged_df_six_month['Station Id'])):
    data = merged_df_six_month['Station Id'][cnt]
    for loc_cnt in range(0,len(location['Station.Id'])):
        loc = location['Station.Id'][loc_cnt]
        if (data - loc) == 0:
            merged_df_six_month['longitude'][cnt] = location['longitude'][loc_cnt]
            merged_df_six_month['latitude'][cnt] = location['latitude'][loc_cnt]   
    loc_cnt = -1
    
cnt += 1
loc_cnt += 1
```
