```python
df1 = pd.read_csv('04JourneyDataExtract01Apr2016-30Apr2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df2 = pd.read_csv('05JourneyDataExtract01May2016-17May2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df3 = pd.read_csv('06JourneyDataExtract18May2016-24May2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df4 = pd.read_csv('07JourneyDataExtract25May2016-31May2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df5 = pd.read_csv('08JourneyDataExtract01Jun2016-07Jun2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df6 = pd.read_csv('09JourneyDataExtract08Jun2016-14Jun2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df7 = pd.read_csv('10JourneyDataExtract15Jun2016-21Jun2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df8 = pd.read_csv('11JourneyDataExtract22Jun2016-28Jun2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
df9 = pd.read_csv('12JourneyDataExtract29Jun2016-05Jul2016.csv',parse_dates = ['Start Date'],infer_datetime_format = True,dayfirst = True)
```

```python
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9]
dataset = pd.concat(frames)
dataset
```

```python
dataset = dataset[(dataset['Start Date'] >= '2016-04-01') & (dataset['Start Date'] <= '2016-06-30')]
dataset.drop(columns = ['Unnamed: 9','Unnamed: 10','Unnamed: 11'], inplace = True)
dataset
```

## Missing values

```python
# Convert the empty string to NaN first
# Using DataFrame.mask() method
newDataset = dataset.mask(dataset == '')

# Number of NaN values 
newDataset.isnull().sum()
```

```python
# Drop all instances with missing values.

newDataset = newDataset.dropna(axis=0, how='any')
newDataset
```

```python
# Convert EndStation Id from float to int
# this has to be done after removing the missing value 
dataset = dataset.fillna(0)
dataset['EndStation Id'] = dataset['EndStation Id'].astype(int)

dataset.dtypes
```

## Extract required information from dataset
### To predict number of bikes in each station in near future
- Number of arrival bike
- Number of depature bike



```python
# Extract the daily data by the date in the dataset.

newDataset = (pd.to_datetime(dataset['End Date'], dayfirst = True)
        .dt.floor('d')
        .reset_index(name = 'timestamp'))
        

newDataset['EndStation Id'] = dataset['EndStation Id'].values
newDataset['StartStation Id'] = dataset['StartStation Id'].values

newDataset
```

```python
# Extract number of bikes arriving to the station
In = newDataset.groupby(['timestamp','EndStation Id'])['EndStation Id'].count().reset_index(name ='In')
In = In[(In['timestamp'] >= '2016-04-01') & ((In['timestamp'] <= '2016-06-30'))]
In
```

```python
# Extract number of bikes leaving to the station
Out = newDataset.groupby(['timestamp','StartStation Id'])['StartStation Id'].count().reset_index(name ='Out')
Out.replace(['1970-01-01'], ['2016-04-01'], inplace = True) 
#Out = Out[(Out['timestamp'] >= '2016-04-01') & ((Out['timestamp'] <= '2016-06-30'))]
Out
```

```python
# combine dataframe : In and Out 

InOut = In[['timestamp', 'In']].copy()
InOut['Out'] = Out[['Out']].copy()
InOut['Station Id'] =  Out[['StartStation Id']].copy()
InOut['Total'] = In['In'] - Out['Out']

InOut = InOut.fillna(0)
InOut['Out'] = InOut['Out'].astype(int)
InOut['Station Id'] = InOut['Station Id'].astype(int)

InOut
```

```python
# To calculate the average values for conditions

num_hours = 24


average_df = df_by_day[['is_holiday', 'is_weekend']].copy()
average_df['avg_temp'] = df_by_day['t1']/num_hours
average_df['avg_temp_feel'] = df_by_day['t2']/num_hours
average_df['avg_humidity'] = df_by_day['hum']/num_hours
average_df['avg_windSpeed'] = df_by_day['wind_speed']/num_hours
average_df['date'] = condition_df['date']
```

### Add Longitude and Latitude for the station location

```python
location = pd.read_csv("Cycle hire docking stations.csv")
```

```python
merged_df['longitude'] = 0
merged_df['latitude'] = 0
```

```python
#merged_df['longitude'] = location['longitude'].apply(lambda x: [location[['longitude']].copy() for y in x])
```

```python
cnt = 0
loc_cnt = 0

# if time permits, try using lambda / apply methods not for loop. 

for cnt in range(0, len(merged_df['Station Id'])):
    data = merged_df['Station Id'][cnt]
    for loc_cnt in range(0,len(location['Station.Id'])):
        loc = location['Station.Id'][loc_cnt]
        if (data - loc) == 0:
            merged_df['longitude'][cnt] = location['longitude'][loc_cnt]
            merged_df['latitude'][cnt] = location['latitude'][loc_cnt]   
    loc_cnt = -1
    
cnt += 1
loc_cnt += 1
```
