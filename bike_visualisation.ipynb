# Marker cluster

```python
# Create base map

London = [51.506949, -0.122876]

map = folium.Map(location = London,
                zoom_start = 12, 
                tiles = "CartoDB positron")
    
for grp_name, df in dataset.groupby('exact_date'):
    feature_group = folium.FeatureGroup(grp_name, show = False)
    for row in df.itertuples():
        _cap = row.In
        _lat = row.latitude
        _lon = row.longitude
        folium.CircleMarker(location = [_lat, _lon], 
                            radius = 9, 
                            popup = "("+ str(_cap) +")" + "  " + str(grp_name),
                            fill_color = color_change(_cap), 
                            color = "gray", 
                            fill_opacity = 0.9).add_to(feature_group)
        #MarkerCluster([(_lat, _lon)]).add_to(feature_group)
        
    feature_group.add_to(map)
folium.LayerControl().add_to(map)
```

```python
#Function to change colors

def color_change(c):
    if(c < 15):
        return('red')
    elif(15 <= c < 30):
        return('orange')
    else:
        return('green')
```
