## Map visualisation

```python
# Import

# data set
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

# visualisation
import folium
from folium.plugins import MarkerCluster
```

### Importing the merged dataset

```python
dataset = pd.read_csv(
    'April_to_June_3_months.csv',
    parse_dates=['timestamp'],
    index_col="timestamp"
)
```

```python
dataset = dataset.drop(["index",'is_holiday','is_weekend'], axis = 1)
dataset = dataset.dropna(axis=0, how='any')
dataset['exact_date'] = dataset.index.astype(str)
dataset
```

```python
location = pd.read_csv("Cycle hire docking stations.csv")
```

### Datasets for each stations

```python
obj = dataset.groupby('Station Id')
```

```python
obj.get_group(1)
```

## Visualisation on map

```python
# add datasets to list

df_map = []
dropped = [35,79,80,137,179,198,205,237,241,259,283,285,304,342,346,369,385,391,406,407,
           413,414,415,416,417,418,422,429,434,438,439,493,497,
           502,525,555,567,575,582,585,791,797,798,799,802]
    
for i in range(1, len(location['Station.Id'])):
    if i in dropped:
        continue  
    df_map.append(obj.get_group(i))

num_station = len(df_map[0])
for i in range(0, num_station):    
    num_station = len(df_map[i])
    df_map[i] = df_map[i][['In','Out','Total','Station Id','date','longitude','latitude']]  
    df_map[i] = df_map[i].drop_duplicates()
    df_map[i]['exact_date'] = df_map[i].index.astype(str)
    df_map[i]['Total'] = df_map[i]['Total'].astype(int)
    
```

```python
df_map[0]
```

```python
#Function to change colors

def color_change(c):
    if(c < 0):
        return('red')
    elif(0 <= c < 30):
        return('orange')
    else:
        return('green')
```

# Circle Marker

```python
%time

# Create base map
London = [51.506949, -0.122876]

map = folium.Map(location = London,
                zoom_start = 12, 
                tiles = "CartoDB positron")
    
for grp_name, df in dataset.groupby('exact_date'):
    feature_group = folium.FeatureGroup(grp_name, show = False)
    for row in df.itertuples():
        _cap = row.Total
        _lat = row.latitude
        _lon = row.longitude
        folium.CircleMarker(location = [_lat, _lon], 
                            radius = 9, 
                            popup = "("+ str(_cap) +")" + "  " + str(grp_name),
                            fill_color = color_change(_cap), 
                            color = "gray", 
                            fill_opacity = 0.9).add_to(feature_group)
        
    feature_group.add_to(map)
folium.LayerControl().add_to(map)
```

## Attempt with station based structure

```python
# Create base map

%time

locations = []
whole_location = []
num_station = len(df_map[0])

for i in range(0,num_station):
    # make latitude and longitude into a list of tuples
    loc = df_map[i][['latitude', 'longitude']].copy()
    locations = list(loc.to_records(index = False))
    whole_location.append(locations[1]) # since all latitude and longitude in df_map[i] are the same
    London = [51.506949, -0.122876]
    map = folium.Map(location = London,
                     zoom_start = 12, 
                     tiles = "CartoDB positron")
    marker_cluster = MarkerCluster(whole_location).add_to(map)
    
num_station = len(df_map[0])

for i in range(1,num_station):
    num_station = len(df_map[i])
    cap = df_map[i]['Total']
    date = df_map[i].index # timestamp
    lat = df_map[i]['latitude']
    lon = df_map[i]['longitude'] 
    
    # Plot markers
    for _lat, _lon, _cap, _date in zip(lat, lon, cap, date):
        folium.CircleMarker(location = [_lat, _lon], 
                            radius = 9, 
                            popup = "("+ str(_cap) +")" + "  " + str(_date),
                            fill_color = color_change(_cap), 
                            color = "gray", 
                            fill_opacity = 0.9).add_to(marker_cluster)
```

```python
dataset = dataset[['In','Out','Total','Station Id','date','longitude','latitude']]  
dataset
```

```python
dataset['exact_date'] = dataset.index.astype(str)
```

```python
# only include locations that exist after data clearance

num_station = len(df_map[0])

for i in range(0,num_station):
    # make latitude and longitude into a list of tuples
    num_station = len(df_map[i])
    loc = df_map[i][['latitude', 'longitude']].copy()
    locations = list(loc.to_records(index = False))
```

```python
# obtain all the locations for the station 

df_location = pd.DataFrame(locations)
df_location.columns = ['location']
df_location['location'] = df_location['location'].drop_duplicates()
df_location = df_location.dropna(axis=0, how ='any')
```

```python
loc = df_location['location'].copy()
all_stations = loc.tolist()
```

```python
%time

# Create base map

London = [51.506949, -0.122876]

map = folium.Map(location = London,
                zoom_start = 12, 
                tiles = "CartoDB positron")
    
for grp_name, df in dataset.groupby('exact_date'):
    feature_group = folium.FeatureGroup(grp_name, show = False)
    make_cluster = MarkerCluster(all_stations).add_to(feature_group)
    for row in df.itertuples():
        _cap = row.Total
        _lat = row.latitude
        _lon = row.longitude
        folium.CircleMarker(location = [_lat, _lon], 
                            radius = 9, 
                            popup = "("+ str(_cap) +")" + "  " + str(grp_name),
                            fill_color = color_change(_cap), 
                            color = "gray", 
                            fill_opacity = 0.9).add_to(make_cluster)
        
    feature_group.add_to(map)
folium.LayerControl().add_to(map)
    
```
